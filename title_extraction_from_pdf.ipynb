```python
from langdetect import detect
from PIL import Image
import pytesseract
import sys
from pdf2image import convert_from_path
import os
from math import sqrt
import cv2
from google.colab.patches import cv2_imshow
import pytesseract
from pytesseract import Output
import cv2
import operator
from PIL import Image
import pytesseract
import sys
from pdf2image import convert_from_path
import os
from iso639 import languages
import fitz
```

```python
PDF_file = "input1.pdf"
pages = convert_from_path(PDF_file, 500)
image_counter = 1

for page in pages:
	filename = "page_"+str(image_counter)+".jpg"
	page.save(filename, 'JPEG')
	image_counter = image_counter + 1
```

```python
pdf_document = "input1.pdf"
doc = fitz.open(pdf_document)
page1 = doc.loadPage(0)
ptext = page1.getText("text")
```

```python
img = cv2.imread('page_1.jpg')

lang = detect(ptext)
xyz = languages.get(alpha2=lang)
l = xyz.bibliographic


d = pytesseract.image_to_data(img, output_type=Output.DICT, lang=l)
text = pytesseract.image_to_string(img, lang=l)
size = []
n_boxes = len(d['level'])
for i in range(n_boxes):
    (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    size.append(h)
final ={}

for i,j in zip(d['text'], size):
  if(i != '' and (text.find(i)!= -1)):
    final[i] = j

final = dict( sorted(final.items(), key=operator.itemgetter(1),reverse=True))
print(final)
h = list(final.values())
c = len(h)-1
w = list(final.keys())
title = w[0]
for k in range(1, c):
  if(h[0]-h[k] < 50):
    title = title + " " + w[k]

para = text.replace(title,"",1).strip("\n")
img = cv2.resize(img, (1000,1000), interpolation=cv2.INTER_LINEAR)
cv2_imshow(img)
```
